```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv("StockMarket.csv")
data['date_time'] = pd.to_datetime(data['date_time'])
```

```python
#data.dtypes
```

```python
fig=plt.figure()
#fig.show()
ax=fig.add_subplot(111)
#ax.plot(data['date_time'],data['alphabet_open'])          
ax.plot(data['date_time'],data['boa_open'], label='BOA')          
ax.plot(data['date_time'],data['gm_open'], label='GM')               
ax.plot(data['date_time'],data['johnson_open'], label='JHN')             
ax.plot(data['date_time'],data['jacobs_open'], label='JCB')           
#ax.plot(data['date_time'],data['tesla_open'])            
ax.plot(data['date_time'],data['albemarle_open'], label='ALB')         
ax.plot(data['date_time'],data['alico_open'], label='ALC')             
ax.plot(data['date_time'],data['disney_open'], label='DSN')            
ax.plot(data['date_time'],data['att_open'], label='ATT') 

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.draw()
```
